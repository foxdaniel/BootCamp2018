**Exercise 8.1**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 100, 1000)
y_1 = (2 / 3) * x + (4 / 3)
y_2 = (1 / 6) * x - (1 / 6)
y_3 = 6 - x

y_max = np.maximum(y_1, y_3)

plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

plt.fill_between(x, y_1, y_max, interpolate=True)
plt.xlim(0,10)
plt.ylim(0,10)
plt.title('Exercise 8.1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)

y = []
for i in range(3):
    y.append((5 / 4) * x - i)

for i in range(3):
    plt.plot(x, y[i], 'k--')
    
plt.fill_between(x, y_1, y_max, interpolate=True)
plt.xlim(0,10)
plt.ylim(0,10)
plt.title('Exercise 8.1')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

The optimizer for this problem is where the two binding constraints intersect; namely, at the point $(x,y)=\bigg(\frac{14}{5}, \frac{16}{5}\bigg)=(2.8, 3.2)$

**Exercise 8.2 (i)**

```python
x = np.linspace(0, 100, 1000)
y_1 = 5 - x / 3
y_2 = 6 - 2 * (x / 3)
y_3 = x - 4
y_4 = np.minimum(y_1, y_2)
y_5 = np.minimum(y_3, y_4)



plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.fill_between(x, 0, y_4, where= x<4, facecolor='blue', interpolate=True)
plt.fill_between(x, y_3, y_4, where= x<6, interpolate=True)
plt.plot(x, 20 - 3 * x)
plt.xlim(0, 10)
plt.ylim(0, 10)
plt.title('Exercise 8.2 (i)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

By checking each of the vertices of the feasible set (blue shaded area), The Fundamental Theorem yields an optimal point at $(x, y)=(6, 2)$. This in turn yields an optimal value of $3(6) + 2 = 20$.

**Exercise 8.2 (ii)**

```python
x = np.linspace(0, 100, 1000)
y_1 = 11 + x
y_2 = 27 - x
y_3 = (90 - 2 * x) / 5
y_4 = np.minimum(y_1, y_3)
y_5 = np.minimum(y_2, y_4)



plt.plot(x, y_1)
plt.plot(x, y_2)
plt.plot(x, y_3)
plt.fill_between(x, y_5, interpolate=True)
plt.plot(x, (132 / 6 - (2 / 3) * x))
plt.xlim(0, 50)
plt.ylim(0, 50)
plt.title('Exercise 8.2 (ii)')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

By checking each of the vertices of the feasible set (blue shaded area), The Fundamental Theorem yields an optimal point at $(x, y)=(15, 12)$. This in turn yields an optimal value of $4(15) + 6(12) = 132$.
