**Problem 1**

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def p1a(n):
    A = np.random.normal(size=(n,n))
    return np.var(np.mean(A, axis=1))
```

```python
def p1b():
        B = np.arange(100, 1000, 100)
        return np.array([p1a(n) for n in B])

plt.plot(p1b())
plt.show()
```

**Problem 2**

```python
def p2():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
    y1 = np.sin(x)
    y2 = np.cos(x)
    y3 = np.arctan(x)

    plt.plot(x, y1, label='sin(x)')
    plt.plot(x, y2, label='cos(x)')
    plt.plot(x, y3, label='arctan(x)')
    plt.legend()
    
    return plt.show()
```

```python
p2()
```

**Problem 3**

```python
def p3():
    x1 = np.linspace(-2, 1, 100)
    x1 = x1[:-1]
    x2 = np.linspace(1, 6, 100)
    x2 = x2[1:]
    y1 = 1 / (x1 - 1)
    y2 = 1 / (x2 - 1)
    
    plt.plot(x1, y1, 'm--', lw=4)
    plt.plot(x2, y2, 'm--', lw=4)
    plt.xlim([-2, 6])
    plt.ylim([-6, 6])
    return plt.show()
```

```python
p3()
```

**Problem 4**

```python
def p4():
    x = np.linspace(0, 2 * np.pi, 100)

    ax1 = plt.subplot(221)
    plt.axis([0, 2 * np.pi, -2, 2])
    plt.title("sin(x)")
    ax1.plot(x, np.sin(x), 'g-')
    ax2 = plt.subplot(222)
    plt.axis([0, 2 * np.pi, -2, 2])
    plt.title("sin(2x)")
    ax2.plot(x, np.sin(2 * x), 'r--')
    ax3 = plt.subplot(223)
    plt.axis([0, 2 * np.pi, -2, 2])
    plt.title("2sin(x)")
    ax3.plot(x, 2 * np.sin(x), 'b--')
    ax4 = plt.subplot(224)
    plt.axis([0, 2 * np.pi, -2, 2])
    plt.title("2sin(2x)")
    ax4.plot(x, 2 * np.sin(2 * x), 'm:')
    
    plt.suptitle('Variations of sin(x)')

    return plt.show()
```

```python
p4()
```

**Problem 5**

```python
def p5():
    fars = np.load('FARS.npy')
    
    fig, axes = plt.subplots(1, 2, figsize=(10, 3))

    axes[0].scatter(fars[:, 1], fars[:, 2], c='k', s=0.5)
    axes[0].axis('equal')

    hours = range(0, 25, 1)
    axes[1].hist(fars[:, 0], bins=hours, edgecolor='k')
    axes[1].set(xlim=(0, 24), xlabel='Hours')

    plt.show()
    
    return plt.show()
```

```python
p5()
```

**Problem 6**

```python
def p6():
    x = np.linspace(-2 * np.pi, 2 * np.pi, 100)
    y = x.copy()
    X, Y = np.meshgrid(x, y)
    Z = (np.sin(X) * np.sin(Y)) / (X * Y)

    plt.subplot(121)
    plt.pcolormesh(X, Y, Z, cmap="jet") #prism looks cool
    plt.colorbar()
    #plt.axis([-2 * np.pi, 2 * np.pi, -2 * np.pi, 2 * np.pi])
    plt.subplot(122)
    plt.contour(X, Y, Z, cmap='prism')
    plt.colorbar()
    
    return plt.show()
```

```python
p6()
```
