```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pydataset import data
```

```python
iris = data('iris')
iris
```

```python
colors = ['purple', 'blue', 'pink']
```

```python
fig, ax = plt.subplots()

for species, color in zip(iris['Species'].unique(), colors):
    iris[iris['Species'] == species].plot(x='Petal.Length', y='Sepal.Length', kind='scatter', 
                                          ax=ax, c=color, label=str(species))
plt.show()
```

```python
fig, ax = plt.subplots()

for species, color in zip(iris['Species'].unique(), colors):
    iris[iris['Species'] == species].plot(x='Petal.Length', y='Sepal.Width', kind='scatter', 
                                          ax=ax, c=color, label=str(species))
plt.show()
```

**Which species is easiest to distinguish from other? How?**
<br>
Setosa is the easiest to distinguish, based on it's characteristics of petal length, sepal width, and sepal length.

```python
diamonds = data('diamonds')
diamonds # Scatter wouldn't be good b/c we want to differentiate by color

# Use bar maybe?
```

```python
diamonds.groupby('color')['price'].mean().plot(kind='bar')
plt.ylabel('price')
plt.show()
```

```python
diamonds.groupby('cut')['price'].mean().plot(kind='bar')
plt.ylabel('price')
plt.show()
```

**How does the color and cut of a diamond affect its price?**
<br>
Diamonds are just expensive no matter what. As we move right on the color
scheme, the price increases. The cut doesn't seem to make a huge impact, but
Premium and Very Good cut are among the top three most expensive

```python
diamonds['dimensions'] = diamonds['x'] * diamonds['y'] * diamonds['z']
diamonds[diamonds.color == 'H'].groupby('cut')['dimensions'].mean().plot(kind=
                                                                      'bar')
plt.show()
```

**Of the diamonds with color "H", those with a "Fair" cut sell, on average, for a
higher price than those with an "Ideal" (superior) cut. What other factors could
explain this unintuitive statistic?**
<br>
It turns out that the variables "x", "y", and "z" in the data correspond to 
the dimensions (length, width, height) of each diamond. From the graph above,
it seems that bigger diamonds are might have lower cut but have higher prices
on average. Shine bright like a diamond!

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
titanic.groupby('Embarked')['Survived'].mean()
```

```python
pd.pivot_table(titanic, values='Survived', index='Embarked', columns='Sex')
```

**What do these tables suggest to you about the significance of where people 
embarked in influencing their survival rate? Examine the context of the 
problem, and explain what you think this really means.**
<br>
There seems to be a correlation between the city from which passengers
embarked from; however, there could be another channel working through ebarkment city that affects the survival rate.

```python
print("\nSurvival rates by embarked, gender, and fare:\n")
fare = pd.qcut(titanic["Fare"], 3)
print(titanic.pivot_table(values="Survived", 
                          index=["Embarked", fare], columns="Sex"))

print("\nSurvival rates by embarked, gender, and age:\n")
age = pd.cut(titanic['Age'], [0, 12, 18, 80])
print(titanic.pivot_table(values="Survived", 
                          index=["Embarked", age], columns="Sex"))
```

What we see here is that females embarking from Q and S had the most 
expensive tickets, and were also less likely to have survived. We also learn
that females in general were more likely to survive. This could be because
women and children might have been able to board the life rafts first before 
the men.
