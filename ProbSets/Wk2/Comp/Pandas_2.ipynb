```python
import seaborn as sbn
from pydataset import data
from matplotlib import pyplot as plt
```

**Arbuthnot**

```python
arbuth = data("Arbuthnot")
arbuth.index= arbuth["Year"]
```

```python
arbuth.plot(y=["Males", "Females"])
plt.ylabel("Births")
plt.title("London Births (Male, Female)")
plt.show()
```

```python
sbn.regplot(x='Year', y='Ratio', data=arbuth)
plt.title("Male to Female Birth Ratio")
plt.show()
```

Source: Arbuthnot, John (1710). "An argument for Devine Providence, taken from the constant Regularity observ'd in the Births of both Sexes," Philosophical transactions, 27, 186-190. Published in 1711.

**Road**

```python
road = data('road')
```

```python
sbn.regplot(x='drivers', y='deaths', data=road)
plt.title("Drivers and Road Deaths")
plt.show()
```

```python
sbn.regplot(x='drivers', y='temp', data=road)
plt.title("Number of Drivers by Temperature")
plt.show()
```

Source: Imperial College, London M.Sc. exercise

**Birthdeathrates**

```python
bth_dth_rates = data('birthdeathrates')
```

```python
sbn.regplot(x='birth', y='death', data=bth_dth_rates)
plt.title("Birth and Death Rates by Country")
plt.show()
```

```python
bth_dth_rates[["birth" , "death"]].hist(grid=False, 
                                        bins=20, edgecolor='k')
plt.xlabel("Rate")
plt.ylabel("Number of Countries")
plt.show()
```

Source: J. A. Hartigan (1975), Clustering Algorithms. John Wiley & Sons, New York.

**Birthwht**

```python
bth_wt = data('birthwt')
```

```python
bth_wt.loc[bth_wt["race"] == 1, "race"] = "White"
bth_wt.loc[bth_wt["race"] == 2, "race"] = "Black"
bth_wt.loc[bth_wt["race"] == 3, "race"] = "Other"
```

```python
sbn.regplot(x='bwt', y='lwt', data=bth_wt)
plt.title("Weight - bwt vs. lwt")
plt.show()
```

```python
race = bth_wt.groupby("race")
bth_wt.boxplot(["bwt"], by="race", grid=False)
plt.title("Birthweight by Race")
plt.ylabel("Weight (grams)")
plt.xlabel("Race")
plt.suptitle("")
plt.show()
```

Source: Hosmer, D.W. and Lemeshow, S. (1989) Applied Logistic Regression. New York: Wiley

**Lung**

```python
lung = data('lung')
```

```python
lung.loc[lung["sex"] == 1, "sex"] = "Male"
lung.loc[lung["sex"] == 2, "sex"] = "Female"
```

```python
lung["age"].hist(grid=False, bins=10, edgecolor='k')
plt.xlabel('Age (yrs)')
plt.ylabel('Distribution')
plt.title('Distribution of Ages')
plt.show()
```

```python
sbn.regplot(x='wt.loss', y='time', data=lung)
plt.title("Weight Loss and Longevity")
plt.xlabel('Weight Loss (past 6 months)')
plt.ylabel('Longevity (days)')
plt.show()
```

Source: Loprinzi CL. Laurie JA. Wieand HS. Krook JE. Novotny PJ. Kugler JW. Bartel J. Law M. Bateman M. Klatt NE. et al. Prospective evaluation of prognostic variables from patient-completed questionnaires. North Central Cancer Treatment Group. Journal of Clinical Oncology. 12(3):601-7, 1994.
