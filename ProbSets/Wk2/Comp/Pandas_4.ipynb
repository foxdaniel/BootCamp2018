**Problem 4**

```python
web_traffic = pd.read_csv('website_traffic.csv')
web_traffic # Checking data
```

```python
web_traffic = pd.read_csv('website_traffic.csv', 
                          parse_dates=['ENTER', 'LEAVE'], 
                          index_col=1) # change to data value

web_traffic['DURATION'] = (web_traffic.LEAVE - web_traffic.index).dt.seconds

fig, ax = plt.subplots()
web_traffic.resample('1min').mean().plot(ax=ax)
web_traffic.resample('H').mean().plot(ax=ax)
plt.legend(['avg visit per min', 'avg visit per hour'])
plt.ylabel('Duration')
plt.show()
```

**Problem 6**

```python
window_span = [30, 120, 365]

plt.plot(djia, label='original')

for i in window_span:
    plt.plot(djia.rolling(window=i).mean(), label=f'window = {i}')
plt.legend()
plt.title("Rolling Average")
plt.show()
```

```python
window_span = [30, 120, 365]

plt.plot(djia, label='original')

for i in window_span:
    plt.plot(djia.ewm(span=i).mean(), label=f'span = {i}')
plt.legend()
plt.title("Rolling Exponential Average")
plt.show()
```

```python
window_span = [30, 120, 365]

plt.plot(djia, label='original')

for i in window_span:
    plt.plot(djia.rolling(window=i).min(), label=f'window = {i}')
plt.legend()
plt.title("Minimum Rolling Values")
plt.show()
```

```python
window_span = [30, 120, 365]

plt.plot(djia, label='original')

for i in window_span:
    plt.plot(djia.rolling(window=i).max(), label=f'window = {i}')
plt.legend()
plt.title("Maximum Rolling Values")
plt.show()
```

**Describe how varying the length of the window changes the approximation of
the data**
<br>
Increasing the length of the window seems to pull the approximation farther
away from the original. That is, if our window size is smaller, then we only consider nearby observations. Hence, we deviate from the data as we increase that window.
